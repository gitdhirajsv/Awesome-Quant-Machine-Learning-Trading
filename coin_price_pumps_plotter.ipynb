```python
!pip install python-binance pandas numpy scikit-learn tensorflow matplotlib

# Import necessary libraries
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from binance.client import Client
from binance.exceptions import BinanceAPIException, BinanceRequestException

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# -------------------------
# 1. Initialize Binance Client
# -------------------------

# If you have API keys, enter them here. Otherwise, set to None for public endpoints.
api_key = 'h0ietEcs4wCTw42cHrgN3aaUwWHIVR9Gz2gAZ5VqyLLMuQIwbr1BJsMFJPFnMsKZ'  # Replace with your API key if available
api_secret = 'sG7uZVcyeeFgdc1SF6VskShQWmeMIWsTiAlU8RF8RHrz02IgAGAPmgahzzwnhZ1v'  # Replace with your API secret if available

client = Client(api_key, api_secret)

# -------------------------
# 2. Define Symbol List
# -------------------------

symbol_pairs = [
    'ZRXUSDT', '1INCHUSDT', 'AAVEUSDT', 'AdExUSDT', 'ACHUSDT', 'ALGOUSDT',
    'TLMUSDT', 'ALPINEUSDT', 'FORTHUSDT', 'ANKRUSDT', 'APEUSDT', 'API3USDT',
    'APTUSDT', 'ARBUSDT', 'ASTRUSDT', 'AUDIOUSDT', 'AVAXUSDT', 'AXLUSDT',
    'AXSUSDT', 'BALUSDT', 'BNTUSDT', 'BANDUSDT', 'BATUSDT', 'BICOUSDT',
    'BTCUSDT', 'BCHUSDT', 'BLURUSDT', 'BNBUSDT', 'BOSONUSDT', 'BTRSTUSDT',
    'ADAUSDC', 'CTSIUSDT', 'CELRUSDT', 'CELOUSDT', 'LINKUSDT', 'CHZUSDT',
    'CLVUSDT', 'COMPUSDT', 'ATOMUSDT', 'COTIUSDT', 'CUDOSUSDT', 'CRVUSDT',
    'DAIUSDT', 'DASHUSDT', 'MANAUSDT', 'DIAUSDT', 'DGBUSDT', 'DOGEUSDT',
    'XECUSDT', 'ENJUSDT', 'EOSUSDT', 'ETHUSDT', 'ETCUSDT', 'ENSUSDT',
    'FTMUSDT', 'PORTOUSDT', 'FETUSDT', 'FILUSDT', 'FLOKIUSDT', 'FLOWUSDT',
    'FLUXUSDT', 'FORTUSDT', 'GUSDT', 'GALAUSDT', 'JAMUSDT', 'GTCUSDT',
    'GLMUSDT', 'ONEUSDT', 'HBARUSDT', 'ZENUSDT', 'ICXUSDT', 'RLCUSDT',
    'ILVUSDT', 'IMXUSDT', 'ICPUSDT', 'IOSTUSDT', 'IOTAUSDT', 'KDAUSDT',
    'KAVAUSDT', 'KSMUSDT', 'KNCUSDT', 'LAZIOUSDT', 'LOKAUSDT', 'LDOUSDT',
    'LSKUSDT', 'LTCUSDT', 'LPTUSDT', 'LOOMUSDT', 'LRCUSDT', 'LTOUSDT',
    'MKRUSDT', 'PONDUSDT', 'MASKUSDT', 'DARUSDT', 'EGLDUSDT', 'MXCUSDT',
    'ALICEUSDT', 'XNOUSDT', 'NEARUSDT', 'NEOUSDT', 'NMRUSDT', 'ROSEUSDT',
    'OCEANUSDT', 'ONGUSDT', 'ONTUSDT', 'OPUSDT', 'ORBSUSDT', 'OXTUSDT',
    'OGNUSDT', 'TRACUSDT', 'PAXGUSDT', 'DOTUSDT', 'MATICUSDT', 'POLYXUSDT',
    'PROMUSDT', 'QTUMUSDT', 'QNTUSDT', 'RADUSDT', 'RVNUSDT', 'REEFUSDT',
    'RENUSDT', 'RENDERUSDT', 'REQUSDT', 'SANTOSUSDT', 'SHIBUSDT',
    'SKLUSDT', 'SLPUSDT', 'SOLUSDC', 'STGUSDT', 'XLMUSDT', 'STMXUSDT',
    'STORJUSDT', 'SUIUSDT', 'RAREUSDT', 'SUSHIUSDT', 'SNXUSDT', 'SYSUSDT',
    'USDTUSD', 'XRPUSDT', 'XTZUSDT', 'GRTUSDT', 'SANDUSDT', 'THETAUSDT',
    'TUSDT', 'UNIUSDT', 'USDCUSDT', 'VETUSDT', 'VTHOUSDT', 'VITEUSDT',
    'VOXELUSDT', 'WAXPUSDT', 'WBTCBTC', 'YFIUSDT', 'ZECUSDT', 'ZILUSDT'
]

# -------------------------
# 3. Fetch Historical Kline Data
# -------------------------

def fetch_historical_klines(symbol, interval, start_str, end_str=None):
    """
    Fetches historical kline data from Binance.

    :param symbol: Cryptocurrency pair symbol.
    :param interval: Kline interval (e.g., '1m', '5m', '1h', '1d').
    :param start_str: Start time in 'YYYY-MM-DD' format.
    :param end_str: End time in 'YYYY-MM-DD' format.
    :return: List of klines.
    """
    try:
        klines = client.get_historical_klines(symbol, interval, start_str, end_str)
        return klines
    except BinanceAPIException as e:
        print(f"Binance API Exception for {symbol}: {e}")
    except BinanceRequestException as e:
        print(f"Binance Request Exception for {symbol}: {e}")
    except Exception as e:
        print(f"General Exception for {symbol}: {e}")
    return None

# -------------------------
# 4. Label Data Based on Price Change
# -------------------------

def label_data(df, threshold=15):
    """
    Labels each 24-hour period as pump (1) if price increased by more than threshold%, else normal (0).

    :param df: DataFrame with 'Open' and 'Close' prices.
    :param threshold: Percentage threshold for labeling as pump.
    :return: DataFrame with an additional 'Pump' column.
    """
    df['Price_Change'] = ((df['Close'] - df['Open']) / df['Open']) * 100
    df['Pump'] = df['Price_Change'].apply(lambda x: 1 if x >= threshold else 0)
    return df

# -------------------------
# 5. Feature Engineering
# -------------------------

def engineer_features(df):
    """
    Engineers features from kline data.

    :param df: DataFrame with kline data.
    :return: DataFrame with engineered features.
    """
    features = pd.DataFrame()
    features['Open'] = df['Open']
    features['High'] = df['High']
    features['Low'] = df['Low']
    features['Close'] = df['Close']
    features['Volume'] = df['Volume']

    # Additional Features
    features['High_Low_Spread'] = features['High'] - features['Low']
    features['Close_Open_Diff'] = features['Close'] - features['Open']
    features['Price_Change'] = ((features['Close'] - features['Open']) / features['Open']) * 100
    features['Volume_Change'] = features['Volume'].pct_change().fillna(0)

    return features

# -------------------------
# 6. Fetch and Prepare Dataset
# -------------------------

def prepare_dataset(symbol, interval, start_date, end_date):
    """
    Fetches kline data, labels it, and engineers features.

    :param symbol: Cryptocurrency pair symbol.
    :param interval: Kline interval.
    :param start_date: Start date in 'YYYY-MM-DD' format.
    :param end_date: End date in 'YYYY-MM-DD' format.
    :return: Tuple of (features DataFrame, labels Series)
    """
    klines = fetch_historical_klines(symbol, interval, start_date, end_date)
    if klines is None:
        return None, None

    # Create DataFrame
    df = pd.DataFrame(klines, columns=[
        'Open Time', 'Open', 'High', 'Low', 'Close', 'Volume',
        'Close Time', 'Quote Asset Volume', 'Number of Trades',
        'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore'
    ])

    # Convert data types
    df['Open'] = df['Open'].astype(float)
    df['High'] = df['High'].astype(float)
    df['Low'] = df['Low'].astype(float)
    df['Close'] = df['Close'].astype(float)
    df['Volume'] = df['Volume'].astype(float)

    # Label the data
    df = label_data(df, threshold=15)

    # Engineer features
    features = engineer_features(df)
    labels = df['Pump']

    return features, labels

# -------------------------
# 7. Prepare Data for All Symbols
# -------------------------

def collect_data(symbol_pairs, interval='1d', start_date='2021-01-01', end_date='2021-12-31'):
    """
    Collects and aggregates data for all symbols.

    :param symbol_pairs: List of cryptocurrency pair symbols.
    :param interval: Kline interval.
    :param start_date: Start date.
    :param end_date: End date.
    :return: Tuple of (features DataFrame, labels Series)
    """
    all_features = []
    all_labels = []

    for symbol in symbol_pairs:
        print(f"Processing {symbol}...")
        features, labels = prepare_dataset(symbol, interval, start_date, end_date)
        if features is not None and labels is not None:
            # Append symbol identifier
            features['Symbol'] = symbol
            all_features.append(features)
            all_labels.append(labels)
        else:
            print(f"Skipping {symbol} due to data fetching issues.")
        # Respect rate limits
        time.sleep(1)  # Sleep for 1 second between requests

    if all_features and all_labels:
        features_df = pd.concat(all_features, ignore_index=True)
        labels_series = pd.concat(all_labels, ignore_index=True)
        return features_df, labels_series
    else:
        print("No data collected.")
        return None, None

# -------------------------
# 8. Train and Evaluate Models
# -------------------------

def train_and_evaluate(X, y):
    """
    Trains SVM and Neural Network models and evaluates their performance.

    :param X: Features DataFrame.
    :param y: Labels Series.
    """
    # Split into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    # -------------------------
    # SVM Model
    # -------------------------
    print("\nTraining Support Vector Machine (SVM) Model...")
    svm_model = SVC(C=1.0, kernel='linear', probability=True)
    svm_model.fit(X_train, y_train)
    y_pred_svm = svm_model.predict(X_test)
    svm_accuracy = accuracy_score(y_test, y_pred_svm)
    print(f"SVM Accuracy: {svm_accuracy * 100:.2f}%")
    print("SVM Classification Report:")
    print(classification_report(y_test, y_pred_svm))
    print("SVM Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred_svm))

    # -------------------------
    # Neural Network Model
    # -------------------------
    print("\nTraining Neural Network Model...")
    nn_model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        Dropout(0.5),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])

    nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model
    history = nn_model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=32,
        validation_data=(X_test, y_test),
        verbose=0  # Set to 1 to see training progress
    )

    # Evaluate the model
    loss, accuracy = nn_model.evaluate(X_test, y_test, verbose=0)
    print(f"Neural Network Accuracy: {accuracy * 100:.2f}%")
    y_pred_nn = (nn_model.predict(X_test) > 0.5).astype("int32")
    print("Neural Network Classification Report:")
    print(classification_report(y_test, y_pred_nn))
    print("Neural Network Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred_nn))

    # -------------------------
    # Plot Training History
    # -------------------------
    plt.figure(figsize=(12, 5))

    # Accuracy Plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Test Accuracy')
    plt.title('Neural Network Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss Plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Test Loss')
    plt.title('Neural Network Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# -------------------------
# 9. Main Execution
# -------------------------

def main():
    # Define parameters
    interval = '1d'  # 1 day interval
    start_date = '2021-01-01'
    end_date = '2021-12-31'

    # Collect data
    features, labels = collect_data(symbol_pairs, interval, start_date, end_date)

    if features is not None and labels is not None:
        print("\nData Collection Completed.")
        print(f"Total Samples: {features.shape[0]}")
        print(f"Number of Features: {features.shape[1]}")
        print(f"Number of Pump Samples: {labels.sum()}")
        print(f"Number of Normal Samples: {len(labels) - labels.sum()}")

        # Encode categorical feature 'Symbol' using one-hot encoding
        features_encoded = pd.get_dummies(features, columns=['Symbol'])

        # Drop rows with missing values if any
        features_encoded = features_encoded.dropna()
        labels = labels.loc[features_encoded.index]

        # Normalize features
        from sklearn.preprocessing import StandardScaler
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(features_encoded)
        X_scaled = pd.DataFrame(X_scaled, columns=features_encoded.columns)

        # Handle class imbalance if necessary
        from sklearn.utils import resample
        df_full = pd.concat([X_scaled, labels], axis=1)
        df_pump = df_full[df_full['Pump'] == 1]
        df_normal = df_full[df_full['Pump'] == 0]

        # Upsample minority class (pump)
        df_pump_upsampled = resample(df_pump,
                                     replace=True,
                                     n_samples=len(df_normal),
                                     random_state=42)

        # Combine
        df_balanced = pd.concat([df_normal, df_pump_upsampled])

        # Shuffle the dataset
        df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

        X = df_balanced.drop('Pump', axis=1)
        y = df_balanced['Pump']

        # Train and evaluate models
        train_and_evaluate(X, y)
    else:
        print("No data to process.")

if __name__ == "__main__":
    main()
# -------------------------
# 9. Visualizing Pump Coins
# -------------------------

def plot_pump_coins(features_df, labels_series):
    """
    Visualizes pump coins by plotting their price movements and marking the pump periods.

    :param features_df: DataFrame containing engineered features including price data.
    :param labels_series: Series containing pump labels (1 for pump, 0 for normal).
    """
    # Combine features and labels
    features_df['Pump'] = labels_series

    # Loop through each symbol
    unique_symbols = features_df['Symbol'].unique()

    for symbol in unique_symbols:
        # Filter data for each symbol
        symbol_data = features_df[features_df['Symbol'] == symbol]
        pump_data = symbol_data[symbol_data['Pump'] == 1]

        # Plot the closing price of the symbol
        plt.figure(figsize=(10, 6))
        plt.plot(symbol_data['Close'], label=f'{symbol} - Close Price')

        # Mark the pumps with green dots
        plt.scatter(pump_data.index, pump_data['Close'], color='green', label='Pump', zorder=5)

        # Add labels and title
        plt.title(f'{symbol} Price Movement with Pumps', fontsize=14)
        plt.xlabel('Time', fontsize=12)
        plt.ylabel('Price (USDT)', fontsize=12)
        plt.legend()
        plt.grid(True)
        plt.show()

# Fetch and Prepare Data (for a small sample for visualization)
features_df, labels_series = collect_data(symbol_pairs[:5], interval='1d', start_date='2021-01-01', end_date='2021-01-31')

if features_df is not None and labels_series is not None:
    plot_pump_coins(features_df, labels_series)
else:
    print("No data available for plotting.")
```
